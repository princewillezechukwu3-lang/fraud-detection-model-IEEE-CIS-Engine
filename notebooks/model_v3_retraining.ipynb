```python
import pandas as pd
import numpy as np
import joblib

x_train = pd.read_parquet('x_lgb_train.parquet')
y_train = pd.read_parquet('y_lgb_train.parquet')
```

```python
x_train['C1_per_Amt'] = x_train['C1'] / (x_train['TransactionAmt'] + 1)
x_train['is_new_user'] = (x_train['C13'] <= 1).astype(int)
```

```python
FEATURE_COLUMNS = x_train.columns.tolist()
```

```python
import lightgbm as lgb
import json

with open('best_params_v3.json', 'r') as f:
    final_params = json.load(f)

final_params['objective'] = 'binary'
final_params['metric'] = 'auc'
final_params['verbosity'] = -1

train_data = lgb.Dataset(x_train, label=y_train)

model_v3 = lgb.train(final_params, train_data, num_boost_round=500)

joblib.dump(model_v3, r"C:\Users\ADMIN\ml projects\fraud detection project\models\fraud_model_v3.pkl", compress=3)
print("✅ Model v3 saved successfully!")
```

```python
old_metadata = joblib.load(r"C:\Users\ADMIN\ml projects\fraud detection project\API\lgb_feature_metadata_v2.joblib")

new_feature_names = x_train.columns.tolist()

new_metadata = {
    "feature_names": new_feature_names,
    "categorical_features": old_metadata['categorical_features'],
    "categorical_mappings": old_metadata['categorical_mappings']
}

joblib.dump(new_metadata, r"C:\Users\ADMIN\ml projects\fraud detection project\API\lgb_feature_metadata_v3.joblib")
```
