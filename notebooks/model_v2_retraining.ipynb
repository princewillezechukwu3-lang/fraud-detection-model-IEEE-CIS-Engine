```python
import pandas as pd
import numpy as np
import joblib

x_train = pd.read_parquet('x_lgb_train.parquet')
y_train = pd.read_parquet('y_lgb_train.parquet')
```

```python
x_train['C1_per_Amt'] = x_train['C1'] / (x_train['TransactionAmt'] + 1)
x_train['is_new_user'] = (x_train['C13'] <= 1).astype(int)
```

```python
FEATURE_COLUMNS = x_train.columns.tolist()
```

```python
import lightgbm as lgb
import joblib

for col in x_train.select_dtypes(include=['object']).columns:
    x_train[col] = x_train[col].astype('category')

lgb_model_v2 = lgb.LGBMClassifier(
    n_estimators=200,      
    learning_rate=0.05,
    num_leaves=31,
    scale_pos_weight=5,  
    random_state=42,
    n_jobs=-1
)

lgb_model_v2.fit(x_train, y_train)

joblib.dump(lgb_model_v2, r"C:\Users\ADMIN\ml projects\fraud detection project\models\fraud_model_v2.pkl", compress=3)
print("✅ Model v2 saved successfully!")
```

```python
old_metadata = joblib.load(r"C:\Users\ADMIN\ml projects\fraud detection project\API\lgb_feature_metadata.joblib")

new_feature_names = x_train.columns.tolist()

new_metadata = {
    "feature_names": new_feature_names,
    "categorical_features": old_metadata['categorical_features'],
    "categorical_mappings": old_metadata['categorical_mappings']
}

joblib.dump(new_metadata, r"C:\Users\ADMIN\ml projects\fraud detection project\API\lgb_feature_metadata_v2.joblib")
```
